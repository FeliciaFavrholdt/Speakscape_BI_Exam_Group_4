```python
import sys
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import string

# Add local utils to path and initialize project folders
sys.path.append(os.path.abspath('.'))
from utils.setup import init_environment
init_environment()
```

```python
# Load the combined TED dataset
df = pd.read_csv("../data/combined_dataset.csv")

# Preview structure
print(f"Shape of dataset: {df.shape}")
df.head()
```

```python
## Clean Transcript Column

We lowercase text, remove punctuation, and strip noise such as bracketed audience reactions (e.g., [Laughter]).
```

```python
def clean_text(text):
    if pd.isnull(text):
        return ""
    text = text.lower()
    text = re.sub(r"\[.*?\]", "", text)
    text = re.sub(r"\s+", " ", text)
    text = text.translate(str.maketrans("", "", string.punctuation))
    return text.strip()

df["transcript_clean"] = df["transcript"].apply(clean_text)
```

Purpose:

Normalizes transcript text for NLP modeling:

Lowercasing

Removing audience reactions

Removing punctuation

Trimming whitespace



```python
## Extract Text Features

We derive word count, character count, and sentence count to use as linguistic features.
```

```python

Purpose:
Describes the simple linguistic features that will be computed in the next cells.
```

```python
df["word_count"] = df["transcript_clean"].apply(lambda x: len(x.split()))
df["char_count"] = df["transcript_clean"].apply(len)
```

```python
Purpose:
Adds numeric features useful for:

Modeling engagement

Visualizing talk length

Comparing styles of popular vs. unpopular talks
```

```python
import nltk
nltk.download("punkt")
from nltk.tokenize import sent_tokenize

df["sentence_count"] = df["transcript"].apply(lambda x: len(sent_tokenize(x)) if pd.notnull(x) else 0)
```

```python
Purpose:
Estimates the number of sentences, which can be used later to calculate:

Sentence length averages

Text complexity metrics

```

```python
## Save Preprocessed Data

This file will be used in Notebook 04 for feature engineering and modeling.
```

```python
output_path = "../data/preprocessed_data.csv"
df.to_csv(output_path, index=False)
print(f"✅ Preprocessed data saved to: {output_path}")
```

```python
## Extract Readability Metrics

We add sentence- and word-level features, and calculate readability scores using standard NLP formulas.
```

```python
# Derived metrics
df["avg_sentence_length"] = df["word_count"] / df["sentence_count"].replace(0, np.nan)
df["avg_word_length"] = df["char_count"] / df["word_count"].replace(0, np.nan)

# Flesch Reading Ease
df["flesch_score"] = 206.835 - (1.015 * df["avg_sentence_length"]) - (84.6 * df["avg_word_length"])

# Flesch-Kincaid Grade Level
df["flesch_kincaid_grade"] = (0.39 * df["avg_sentence_length"]) + (11.8 * df["avg_word_length"]) - 15.59
```

```python
## Distribution of Linguistic Features

The following histograms show how sentence count, sentence length, and readability scores are distributed across the TED talks.
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
sns.histplot(df["sentence_count"], bins=50, ax=axes[0, 0], color="skyblue")
axes[0, 0].set_title("Sentence Count")

sns.histplot(df["avg_sentence_length"].dropna(), bins=50, ax=axes[0, 1], color="salmon")
axes[0, 1].set_title("Average Sentence Length")

sns.histplot(df["avg_word_length"].dropna(), bins=50, ax=axes[1, 0], color="lightgreen")
axes[1, 0].set_title("Average Word Length")

sns.histplot(df["flesch_score"].dropna(), bins=50, ax=axes[1, 1], color="orchid")
axes[1, 1].set_title("Flesch Reading Ease Score")

plt.tight_layout()
plt.savefig("../plots/readability_distributions.png")
plt.show()
```

## 3. Transcript Recovery and Cleaning Summary

In the TED_2017 dataset, 86 talks were originally missing transcripts. Through recovery efforts, we restored 37 of these using URL cross-referencing and cached archives.

This section visualizes the impact of this recovery on dataset completeness.


```python
# Recovery numbers
original_total = 2549
missing_initial = 86
recovered = 37
excluded = missing_initial - recovered
final_total = original_total - excluded

# Bar chart
labels = ['Original Records', 'Missing Initially', 'Recovered', 'Excluded', 'Final Cleaned']
values = [original_total, missing_initial, recovered, excluded, final_total]

plt.figure(figsize=(10, 5))
bars = plt.bar(labels, values, color=['#9ecae1', '#fcbba1', '#a1d99b', '#fdd0a2', '#bcbddc'])

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 5, f'{height}', ha='center')

plt.title("TED_2017 Transcript Recovery Summary")
plt.ylabel("Record Count")
plt.tight_layout()
plt.savefig("../plots/ted_2017_transcript_recovery.png")
plt.show()
```

## 5. Filtering Short or Empty Transcripts

We remove any transcripts that have fewer than 30 words. This threshold ensures we only retain talks with enough content for meaningful analysis.


```python
# Filter transcripts
initial_count = df.shape[0]
min_words = 30
df_cleaned = df[df['word_count'] >= min_words].copy()
final_count = df_cleaned.shape[0]
removed_count = initial_count - final_count

print(f"Initial records: {initial_count}")
print(f"Removed (too short): {removed_count}")
print(f"Remaining after filtering: {final_count}")
```
