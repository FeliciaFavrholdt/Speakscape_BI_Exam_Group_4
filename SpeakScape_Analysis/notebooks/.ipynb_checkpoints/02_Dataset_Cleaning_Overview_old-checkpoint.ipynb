```python
import matplotlib.pyplot as plt

# Data
recovered = 37
not_recovered = 49

# Plot
plt.figure(figsize=(6, 6))
plt.pie([recovered, not_recovered],
        labels=["Recovered", "Not Recovered"],
        autopct="%1.1f%%",
        colors=["#66c2a5", "#fc8d62"],
        startangle=140)
plt.title("TED_2017 Transcript Recovery Breakdown")
plt.tight_layout()
plt.savefig("../plots/ted_2017_recovery_pie.png")
plt.show()
```
