```python
import shap
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load previously trained model and data
import joblib
model = joblib.load("../models/random_forest_model.pkl")  
X = pd.read_csv("../data/scaled_features.csv")

# Subsample for faster SHAP computation
X_sample = X.sample(n=300, random_state=42)

# Initialize SHAP TreeExplainer
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X_sample)
```

```python
# Plot average magnitude of feature influence for class 1 (high engagement)
shap.summary_plot(shap_values[1], X_sample, plot_type="bar")
```

```python
shap.summary_plot(shap_values, X_sample)
```

## SHAP Analysis: Interpreting Model Predictions

To explain the predictions made by our Random Forest classifier, we used SHAP (SHapley Additive exPlanations). SHAP helps identify which linguistic features most influence whether a talk is predicted to be high or low engagement.

### Why SHAP?
This aligns with our core research goal:
> *What linguistic features most impact success, and how can we make them visible?*

SHAP offers interpretable, per-sample insights — ideal for user-facing feedback systems.

---

### SHAP Summary Plot Interpretation

Each dot represents a talk in the dataset.

- **Y-axis**: Features ranked by importance (top = most influential)
- **X-axis (SHAP value)**: Direction and magnitude of influence on the prediction
  - Positive values → push prediction toward high engagement
  - Negative values → pull prediction toward low engagement
- **Color**: Indicates feature value
  - **Red** = high value of the feature
  - **Blue** = low value of the feature

---

### Key Observations

- **`word_count`**: High word counts tend to increase engagement predictions.
- **`avg_sentence_length`**: Longer average sentence length has a complex effect, sometimes increasing and sometimes decreasing engagement depending on context.
- **`lexical_diversity`**: Greater diversity generally increases the likelihood of high engagement.
- **`duration`** and **`avg_word_length`** also show strong influence on predictions.

These insights support future personalization — we can tell a user *what to adjust* in their speech content to align with engaging TED Talks.
